```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
scores = np.load('set1_"a prime number whose last digit is 7 (e.g., 7, 17, 37, 47, 67, 97)"_20_runs_7_attempts_range_100_200_delete_0_list_1.npy')
#scores = np.load('set1_"a prime number whose last digit is 7 (e.g., 7, 17, 37, 47, 67, 97)"_20_runs_7_attempts_range_100_200_delete_1_list_1.npy')
scores = np.load('set1_"a prime number whose sum of digits is divisible by 7 (e.g., 7, 43, 59, 61)"_20_runs_7_attempts_range_100_200_delete_0_list_1.npy')
scores = np.load('set1_"a prime number whose sum of digits is divisible by 7 (e.g., 7, 43, 59, 61)"_20_runs_7_attempts_range_100_200_delete_1_list_1.npy')
ERRORS = ['Not ok', 'Missing', 'Wrong order', 'Out of range']
N_RUNS, N_ATTEMPTS, N_ERRORS = scores.shape
```

```python
N_RUNS = 20
N_ATTEMPTS = 7
DEFS = ['a prime number whose last digit is 7 (e.g., 7, 17, 37, 47, 67, 97)', 'a prime number whose sum of digits is divisible by 7 (e.g., 7, 43, 59, 61)']
plt.figure(figsize=(10, 30))
k = 0
for START, END in [(100, 200), (100, 400), (200, 100), (400, 100)]:
    for WITH_LIST in [True, False]:
        for DELETE_MESSAGES in [True, False]:
            for iDEF in range(2):
                DEF = DEFS[iDEF]
                file_name = f'set1_"{DEF}"_{N_RUNS}_runs_{N_ATTEMPTS}_attempts_range_{START}_{END}_delete_{int(DELETE_MESSAGES)}_list_{int(WITH_LIST)}.npy'
                k += 1
                try:
                    scores = np.load(file_name)
                except:
                    continue
                print(START, END)
                plt.subplot(8, 4, k)
                plt.plot(np.mean(scores, axis=0))
                if k == 1:
                    plt.legend(ERRORS)
                if k % 4 == 1:
                    plt.ylabel('Number of errors');
                plt.title(iDEF)
                if k >= 13:
                    plt.xlabel('Attempt')
                    
```

```python
plt.figure(figsize=(10, 20))
for j in range(N_ERRORS):
    plt.subplot(4, 1, j+1)
    plt.title(ERRORS[j])
    for i in range(N_RUNS):
        plt.plot(range(N_ATTEMPTS), scores[i, :, j])
        plt.xlabel('Attempt')
        plt.ylabel('Number of errors')
```
